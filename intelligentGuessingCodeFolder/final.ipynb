```python
def isSymbol(ch):
   patternRegex = re.compile(r'[^a-z0-9<>]')
   data =  re.findall(patternRegex, ch)
   if len(data) == 0:
      return False
   return True

def firstNL(inWord, chars):
   if inWord == "" or chars == "":
      return 0
   if inWord[0] == chars[0]:
      return 1+firstNL(inWord[1:],chars[1:])
   return 0

allFNames = [re.sub(r"[^a-z ]", "", "".join(x.lower().split(" "))) for x in dataset["firstname"].to_numpy() if type(x) != float]
allLNames = [re.sub(r"[^a-z ]", "", "".join(x.lower().split(" "))) for x in dataset["lastname"].to_numpy() if type(x) != float]
def getPattern(data, fname, lname):
   parsed = ""
   symbol = ""
   if data == "":
      return ""
   if isSymbol(data[0]):
      return data[0] + getPattern(data[1:], fname,lname)
   if fname != "" and data.startswith(fname):
      parsed = fname
      symbol = "<11>"
   elif lname != "" and data.startswith("".join(lname.split(" "))):
      symbol = "<22>"
      parsed = "".join(lname.split(" "))
   elif fname == "" or lname == "":
      fnameParsed = [s for s in allFNames if data.startswith(s)] 
      lnameParsed = [s for s in allLNames if data.startswith(s)]
      if len(lnameParsed) > 0:
         parsed=lnameParsed[0]
         symbol = '<22>'
      elif len(fnameParsed) > 0:
         parsed=fnameParsed[0]
         symbol = '<11>'
   elif fname != "" and data.startswith(fname[0]):
      symbol = "<1"
      fnl = firstNL(fname, data)
      if fnl == 1:
         symbol += ">"
      else:
         symbol += "0-f"+str(fnl)+"l>"
      parsed = data[0:fnl]
   elif lname !="" and len(lname.split(" ")) > 1 and data.startswith(lname.split(" ")[0]):
      symbol = "<20>"
      parsed = lname.split(" ")[0]
   elif lname !="" and len(lname.split(" ")) > 1 and data.startswith(lname.split(" ")[1]):
      symbol = "<21>"
      parsed = lname.split(" ")[1]
   elif lname !="" and len(lname.split(" ")) == 1 and data.startswith(lname[0]):
      symbol = "<2"
      fnl = firstNL(lname, data)
      if fnl == 1:
         symbol += ">"
      else:
         symbol += "0-f"+str(fnl)+"l>"
      parsed = data[0:fnl]
   elif lname !="" and len(lname.split(" ")) > 1 and data.startswith(lname.split(" ")[0][0]):
      symbol = "<20"
      fnl = firstNL(lname.split(" ")[0], data)
      symbol += "-f"+str(fnl)+"l>"
      parsed = data[0:fnl]
   elif lname !="" and len(lname.split(" ")) > 1 and data.startswith(lname.split(" ")[1][0]):
      symbol = "<21"
      fnl = firstNL(lname.split(" ")[1], data)
      symbol += "-f"+str(fnl)+"l>"
      parsed = data[0:fnl]
   else:
      parsed = 'M'
      symbol='M'
   return symbol + getPattern(data[len(parsed):], fname, lname)
```
